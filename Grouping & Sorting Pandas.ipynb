Select Wines by the country and province

```python
wines_by_country_and_province = reviews.groupby(['country','province']).apply(lambda df: df.title.iloc[0])
```

```python
wines_by_country_and_province
```

```python
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
reviews.groupby(['country', 'province']).apply(lambda df: df.title.loc[df.points.idxmax()])
```
